```python
data = df.dropna()
data.index = pd.to_datetime(data.Date)
data=data['Prev Close']['2019-01-01':'2019-12-01']
data.describe()
```

```python
data.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(data);
```

```python
rollmean = data.rolling(12).mean()
rollstd = data.rolling(12).std()
print(rollmean.head(20))
print(rollstd.head(20))
```

```python
plt.figure(figsize=(16,7))
plt.plot(data, label='Original')
plt.plot(rollmean, label='Rolling Mean')
plt.plot(rollstd, label = 'Rolling Std')
plt.legend()
plt.title('Rolling Mean & Standard Deviation')
plt.show()
```

```python
plt.figure(figsize=(16,7))
ts_log = np.log(data)
plt.plot(ts_log);
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts_log, freq=1, model='multiplicative')
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
plt.figure(figsize=(15,8))
fig = plt.figure(1)

plt.subplot(411)
plt.plot(ts_log, label = 'Original')
plt.legend()
plt.subplot(412)
plt.plot(trend, label = 'Trend')
plt.legend()
plt.subplot(413)
plt.plot(seasonal, label = 'Seasonal')
plt.legend()
plt.subplot(414)
plt.plot(residual, label = 'Residual')
plt.legend()
plt.show()
```

```python
ts_log_diff = ts_log-ts_log.shift(1)
ts_log_diff_mean = ts_log_diff.rolling(12).mean()
ts_log_diff_std = ts_log_diff.rolling(12).std()
plt.figure(figsize=(15,7))
plt.plot(ts_log_diff, label = 'Original')
plt.legend()
plt.plot(ts_log_diff_mean, label = 'Mean')
plt.legend()
plt.plot(ts_log_diff_std, label = 'Standard Deviation')
plt.legend()
plt.title('Rolling Mean & Standard Deviation')
plt.show()
```

```python
data.sort_index(inplace= True)
```

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(ts_log_diff, nlags=20)
lag_pacf = pacf(ts_log_diff, nlags=20)
```

```python
import statsmodels.api as sm
fig = plt.figure(figsize=(12,7))
f1 =fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts_log_diff.dropna(), lags=40, ax=f1)
f2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts_log_diff.dropna(), lags=40, ax=f2 )
```

```python
ts_log_diff=ts_log_diff[~ts_log_diff.isnull()]
ts_log_diff.isna().sum()
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(ts_log_diff, order=(2,1,1))
model_fit = model.fit()

plt.figure(figsize=(12,5))
plt.plot(ts_log_diff, label='Actual')
plt.plot(model_fit.fittedvalues, color='red', label = 'Predicted')
plt.legend()
plt.show()
```

```python
ARIMA_diff_predictions = pd.Series(model_fit.fittedvalues, copy=True)
print(ARIMA_diff_predictions.head())
```

```python
ARIMA_diff_predictions_cumsum = ARIMA_diff_predictions.cumsum()
print(ARIMA_diff_predictions_cumsum.head())
```

```python
ts_log.iloc[1]
```

```python
ARIMA_log_prediction = pd.Series(ts_log.iloc[0], index=ts_log.index)
ARIMA_log_prediction = ARIMA_log_prediction.add(ARIMA_diff_predictions_cumsum,fill_value=0)
ARIMA_log_prediction.head(10)
```

```python
plt.figure(figsize=(12,8))
predictions_ARIMA = np.exp(ARIMA_log_prediction)
plt.plot(data)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-data)**2)/len(data)))
```
